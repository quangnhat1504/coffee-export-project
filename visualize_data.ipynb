## Load and prepare data coffee.csv

### Subtask:
Load the first CSV file into a pandas DataFrame, clean the data by handling missing values and converting data types, and prepare it for analysis.


**Reasoning**:
Import pandas, load the data, handle missing values, and convert data types.



```python
import pandas as pd

# Read the CSV file
df_coffee = pd.read_csv('Data_coffee.csv', encoding='utf-8')

# Handle missing values ('null' and 'Null')
df_coffee.replace(['null', 'Null'], pd.NA, inplace=True)

# Convert data types - attempt to convert all columns except 'Hang_muc' to numeric
for col in df_coffee.columns:
    if col != 'Hang_muc':
        df_coffee[col] = pd.to_numeric(df_coffee[col], errors='coerce')

# Display the first few rows and data types to verify
display(df_coffee.head())
display(df_coffee.info())
```

```python
# Filter out rows where 'Hang_muc' is NaN or 'Hang_muc' is one of the year rows
df_coffee_cleaned = df_coffee.dropna(subset=['Hang_muc'])
df_coffee_cleaned = df_coffee_cleaned[df_coffee_cleaned['Hang_muc'].str.contains('2005') == False]

# Melt the DataFrame to long format
df_coffee_melted = df_coffee_cleaned.melt(id_vars='Hang_muc', var_name='Year', value_name='Value')

# Convert 'Year' to numeric
df_coffee_melted['Year'] = pd.to_numeric(df_coffee_melted['Year'])

# Pivot the table to have metrics as columns
df_coffee_pivot = df_coffee_melted.pivot_table(index='Year', columns='Hang_muc', values='Value')

# Reset index to make 'Year' a column
df_coffee_pivot = df_coffee_pivot.reset_index()

# Rename columns for clarity - removing units from names for easier access
df_coffee_pivot.columns.name = None
df_coffee_pivot = df_coffee_pivot.rename(columns={
    'Area (Thousand ha)': 'Area',
    'San luong ca phe san xuat': 'Production (tons)',
    'San luong ca phe xuat khau (Thousand tons)': 'Exports (tons)',
    'Tong_luong_mua(mm)': 'Rainfall (mm)',
    'Do_am_trung_binh(%)': 'Average Humidity (%)',
    'Nhiet_do_trung_binh(Celcius)': 'Average Temperature (Celcius)',
    'Kim Ngach': 'Trade Value (million USD)',
    'coffee_price_usd_per_ton(world)(USD/ton)': 'World Coffee Price (USD/ton)',
    'coffee_price_usd_per_ton(vietnam)(USD/ton)': 'Vietnam Coffee Price (USD/ton)'
})


display(df_coffee_pivot.head())
display(df_coffee_pivot.info())
```

**Reasoning**:
Remove the redundant 'Hang_muc' column which contains year values after pivoting, as 'Year' column is already present.



```python
# Drop the redundant 'Hang_muc' column
df_coffee_pivot = df_coffee_pivot.drop(columns=['Hang_muc'])

# Display the cleaned DataFrame
display(df_coffee_pivot.head())
display(df_coffee_pivot.info())
```

## Load and prepare thi phan 3 thi truong chinh.csv

### Subtask:
Load the second CSV file into a pandas DataFrame, clean the data by handling missing values and converting data types, and prepare it for analysis.


**Reasoning**:
Load the second CSV file, handle missing values, convert data types, and display the head and info of the resulting DataFrame.



```python
# Read the CSV file
df_thi_phan = pd.read_csv('Thi_phan_3_thi_truong_chinh.csv')

# Handle missing values ('Null') by replacing them with pd.NA
df_thi_phan.replace('Null', pd.NA, inplace=True)

# Convert data types
df_thi_phan['Year'] = pd.to_numeric(df_thi_phan['Year'])
df_thi_phan['Trade Value(million_USD)'] = pd.to_numeric(df_thi_phan['Trade Value(million_USD)'], errors='coerce')
df_thi_phan['Quantity(tons)'] = pd.to_numeric(df_thi_phan['Quantity(tons)'], errors='coerce')

# Display the first few rows and data types to verify
display(df_thi_phan.head())
display(df_thi_phan.info())
```

## Analyze and visualize df coffee pivot

### Subtask:
Analyze the correlations between different variables in the `df_coffee_pivot` DataFrame and create visualizations to show these relationships. This will likely involve creating scatter plots, heatmaps, and other relevant plots to explore how different factors like area, production, and climate influence trade value and prices.


**Reasoning**:
Calculate and visualize the correlation matrix for the `df_coffee_pivot` DataFrame.



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
correlation_matrix = df_coffee_pivot.corr()

# Display the correlation matrix
display(correlation_matrix)

# Create a heatmap of the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Coffee Data Variables')
plt.show()
```

**Reasoning**:
Generate scatter plots to visualize the relationships between key variables as requested in the instructions.



```python
# Generate scatter plots
plt.figure(figsize=(15, 5))

# Scatter plot: Production vs Trade Value
plt.subplot(1, 3, 1)
sns.scatterplot(data=df_coffee_pivot, x='Production (tons)', y='Trade Value (million USD)')
plt.title('Production vs Trade Value')
plt.xlabel('Production (tons)')
plt.ylabel('Trade Value (million USD)')

# Scatter plot: Area vs Production
plt.subplot(1, 3, 2)
sns.scatterplot(data=df_coffee_pivot, x='Area', y='Production (tons)')
plt.title('Area vs Production')
plt.xlabel('Area (Thousand ha)')
plt.ylabel('Production (tons)')

# Scatter plot: Average Temperature vs Production
plt.subplot(1, 3, 3)
sns.scatterplot(data=df_coffee_pivot, x='Average Temperature (Celcius)', y='Production (tons)')
plt.title('Average Temperature vs Production')
plt.xlabel('Average Temperature (Celcius)')
plt.ylabel('Production (tons)')

plt.tight_layout()
plt.show()
```

## Analyze and visualize df thi phan

Create line plots to visualize the trend of trade value and quantity for each importer over time.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot Trade Value over time for each importer
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_thi_phan, x='Year', y='Trade Value(million_USD)', hue='Importer')
plt.title('Trade Value (million USD) Over Time by Importer')
plt.xlabel('Year')
plt.ylabel('Trade Value (million USD)')
plt.grid(True)
plt.show()

# Plot Quantity over time for each importer
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_thi_phan, x='Year', y='Quantity(tons)', hue='Importer')
plt.title('Quantity (tons) Over Time by Importer')
plt.xlabel('Year')
plt.ylabel('Quantity (tons)')
plt.grid(True)
plt.show()
```
