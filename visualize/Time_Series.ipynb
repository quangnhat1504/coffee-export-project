```python
import pandas as pd
import matplotlib.pyplot as plt

# c file
filepath = r"D:/formatted_data.csv"
df = pd.read_csv(filepath)

# Danh sch cc ct (tr ct nm)
attributes = [col for col in df.columns if col != "Year"]

# V biu  cho tng thuc tnh
for col in attributes:
    plt.figure(figsize=(8, 4))
    plt.plot(df["Year"], df[col], marker='o', linestyle='-', linewidth=2)
    plt.title(f"Biu  Time-Series: {col}", fontsize=14)
    plt.xlabel("Nm", fontsize=12)
    plt.ylabel(col, fontsize=11)
    plt.grid(True, linestyle='--', alpha=0.6)
    
    #  t trc X ch hin th s nguyn
    plt.xticks(df["Year"].astype(int), rotation=45)
    
    plt.tight_layout()
    plt.show()

```
