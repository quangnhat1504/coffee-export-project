China 2005-2023

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# Tạo danh sách chứa dữ liệu của tất cả các năm
all_data = []

# Giả lập trình duyệt thật
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/122.0.0.0 Safari/537.36"
}

# Vòng lặp qua các năm 2005–2023
for year in range(2005, 2024):
    url = f"https://wits.worldbank.org/trade/comtrade/en/country/CHN/year/{year}/tradeflow/Imports/partner/VNM/product/090111"
    print(f"\nĐang xử lý năm {year}...")

    try:
        response = requests.get(url, headers=headers, timeout=10)
        print("Status code:", response.status_code)

        if response.status_code != 200:
            print(f"Không tải được dữ liệu năm {year}")
            continue

        soup = BeautifulSoup(response.text, "html.parser")

        # Tìm bảng dữ liệu
        table = soup.find("table")
        if not table:
            print(f"Không thấy bảng dữ liệu năm {year} (trang JS render).")
            continue

        rows = table.find_all("tr")

        # Chuyển các hàng trong bảng thành list
        for row in rows:
            cols = [c.text.strip() for c in row.find_all(["td", "th"])]
            if cols:
                all_data.append([year] + cols)

        print(f"Đã lấy {len(rows)} dòng dữ liệu cho năm {year}.")
        time.sleep(1.5)  # Nghỉ 1.5s tránh bị chặn

    except Exception as e:
        print(f"Lỗi khi xử lý năm {year}: {e}")

# Nếu có dữ liệu, lưu ra CSV
if all_data:
    df = pd.DataFrame(all_data)
    df.to_csv("wits_data_2005_2023.csv", index=False, encoding="utf-8-sig")
    print("\nĐã lưu toàn bộ dữ liệu vào file wits_data_2005_2023.csv")
else:
    print("\nKhông lấy được dữ liệu nào (trang tải bằng JavaScript).")
```

```python
from google.colab import files
import pandas as pd

# Upload từng file
uploaded = files.upload()
# Sau khi upload, bạn sẽ thấy danh sách file trong 'uploaded.keys()'

# Đọc file CSV đã upload
china = pd.read_csv("China 2005-2023.csv")
europe = pd.read_csv("European 2005-2023.csv")
usa = pd.read_csv("United States 2005-2023.csv")

# Gộp các file lại
merged_df = pd.concat([china, europe, usa], ignore_index=True)

# Xóa trùng lặp
merged_df = merged_df.drop_duplicates()

# Xuất file mới
merged_df.to_csv("Merged_Coffee_Trade_2005_2023.csv", index=False)

# Tải về máy
files.download("Merged_Coffee_Trade_2005_2023.csv")
```
