China 2005-2023

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# Táº¡o danh sÃ¡ch chá»©a dá»¯ liá»‡u cá»§a táº¥t cáº£ cÃ¡c nÄƒm
all_data = []

# Giáº£ láº­p trÃ¬nh duyá»‡t tháº­t
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/122.0.0.0 Safari/537.36"
}

# VÃ²ng láº·p qua cÃ¡c nÄƒm 2005â€“2023
for year in range(2005, 2024):
    url = f"https://wits.worldbank.org/trade/comtrade/en/country/CHN/year/{year}/tradeflow/Imports/partner/VNM/product/090111"
    print(f"\nÄang xá»­ lÃ½ nÄƒm {year}...")

    try:
        response = requests.get(url, headers=headers, timeout=10)
        print("Status code:", response.status_code)

        if response.status_code != 200:
            print(f"KhÃ´ng táº£i Ä‘Æ°á»£c dá»¯ liá»‡u nÄƒm {year}")
            continue

        soup = BeautifulSoup(response.text, "html.parser")

        # TÃ¬m báº£ng dá»¯ liá»‡u
        table = soup.find("table")
        if not table:
            print(f"KhÃ´ng tháº¥y báº£ng dá»¯ liá»‡u nÄƒm {year} (trang JS render).")
            continue

        rows = table.find_all("tr")

        # Chuyá»ƒn cÃ¡c hÃ ng trong báº£ng thÃ nh list
        for row in rows:
            cols = [c.text.strip() for c in row.find_all(["td", "th"])]
            if cols:
                all_data.append([year] + cols)

        print(f"ÄÃ£ láº¥y {len(rows)} dÃ²ng dá»¯ liá»‡u cho nÄƒm {year}.")
        time.sleep(1.5)  # Nghá»‰ 1.5s trÃ¡nh bá»‹ cháº·n

    except Exception as e:
        print(f"Lá»—i khi xá»­ lÃ½ nÄƒm {year}: {e}")

# Náº¿u cÃ³ dá»¯ liá»‡u, lÆ°u ra CSV
if all_data:
    df = pd.DataFrame(all_data)
    df.to_csv("wits_data_2005_2023.csv", index=False, encoding="utf-8-sig")
    print("\nÄÃ£ lÆ°u toÃ n bá»™ dá»¯ liá»‡u vÃ o file wits_data_2005_2023.csv")
else:
    print("\nKhÃ´ng láº¥y Ä‘Æ°á»£c dá»¯ liá»‡u nÃ o (trang táº£i báº±ng JavaScript).")
```
