```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc file
df = pd.read_csv(r"D:/formatted_data.csv")

# Danh sách các cột (trừ cột năm)
attributes = [col for col in df.columns if col != "Year"]

# Vẽ biểu đồ cho từng thuộc tính
for col in attributes:
    plt.figure(figsize=(8, 4))
    plt.plot(df["Year"], df[col], marker='o', linestyle='-', linewidth=2)
    plt.title(f"Biểu đồ Time-Series: {col}", fontsize=14)
    plt.xlabel("Năm", fontsize=12)
    plt.ylabel(col, fontsize=11)
    plt.grid(True, linestyle='--', alpha=0.6)
    
    # 🔹 Đặt trục X chỉ hiển thị số nguyên
    plt.xticks(df["Year"].astype(int), rotation=45)
    
    plt.tight_layout()
    plt.show()

```
